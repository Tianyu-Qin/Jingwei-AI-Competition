```python
import cv2
from PIL import Image
import gdal
```

```python
n='2'
ds = gdal.Open('jingwei_round1_train_20190619/image_' + n + '.png')
dslb = gdal.Open('jingwei_round1_train_20190619/image_' + n + '_label.png')
```

```python
wx=ds.RasterXSize
wy=ds.RasterYSize
print(wx)
print(wy)
```

```python
stx = 0
sty = 0
step = 300
outsize = 600
nullthresh = outsize*outsize*0.7
cx = 0
cy = 0
while cy+outsize < wy:
    cx = 0
    while cx+outsize < wx:
        img = ds.ReadAsArray(cx,cy,outsize,outsize) # 4x600x600, delete the 4th channel
        img2 = img[:3,:,:].transpose(1,2,0) # 600x600x3, transpose to RGB
        if (img2[:,:,0]==0).sum() > nullthresh:
            cx += step
            print('kongbai...', cx, cy)
            continue

        img2 = Image.fromarray(img2,'RGB')
        img2.save('./data/train/data600/'+n+'_{}_{}.bmp'.format(cx, cy))
        #deal with label
        img = dslb.ReadAsArray(cx,cy,outsize,outsize)
        img = Image.fromarray(img).convert('L')
        img.save('./data/train/label600/'+n+'_{}_{}.bmp'.format(cx, cy))
 
        cx += step
    cy += step
```
